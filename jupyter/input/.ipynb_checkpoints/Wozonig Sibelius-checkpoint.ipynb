```python
from __future__ import print_function

__author__ = "Martin Aigner"

import requests
import json

class Login:
    def __init__(self, endpoint, access_token, verify_certificate):
        self.endpoint = endpoint
        self.access_token = access_token
        self.verify_certificate = verify_certificate
        self.headers = {
            'User-Agent': 'oma.py',
            'Authorization': "Bearer " + access_token
        }
        self.config = {
            "api_path": "api/",
            "cache_dir": "omacache"
        }
        
    def recording(self, id):
        r = Recording(id, self)
        r.get()
        return r
    
    def composer(self, id):
        c = Composer(id, self)
        c.get()
        return c
        

        
class Recording:
    __data_cache = None
    def __init__(self, id, oma_login):
        self.id = id
        self.oma_login = oma_login
        
    def get(self):
        params = {
            "method": "get",
            "recording": self.id
        }
        url = self.oma_login.endpoint + self.oma_login.config["api_path"] + "recording"
        response = requests.post(url, headers=self.oma_login.headers, data=params, verify=self.oma_login.verify_certificate)
        # print(response.content.decode())
        content = json.loads(response.content.decode())
        self.__data_cache = content
    
    def show(self):
        if self.__data_cache is None:
            self.__get()        
        return self.__data_cache
        
        
class Composer:
    __data_cache = None
    def __init__(self, id, oma_login):
        self.id = id
        self.oma_login = oma_login

        
    def get(self):
        params = {
            "method": "get",
            "composer": self.id
        }
        url = self.oma_login.endpoint + self.oma_login.config["api_path"] + "composer"
        response = requests.post(url, headers=self.oma_login.headers, data=params, verify=self.oma_login.verify_certificate)
        content = json.loads(response.content.decode())
        self.__data_cache = content
        
        
    def show(self):
        if self.__data_cache is None:
            self.__get()
        return self.__data_cache


# login to the Open Music Annotations cloud

# url = "https://oma.digital/"
url = "http://localhost:8080/"
oma_login = Login(url, "3iispvmrt8s69cp97a80tnk3kqdpean8", True)


recording_ids = [147620, 148717, 152399, 138848, 153497, 139946, 144330, 151301, 157532]
recordings = []
for i in recording_ids:
    r = oma_login.recording(i)
    recordings.append(r.show())
```

```python
# using data frames to handle tables and stuff
import pandas as pd

pd.DataFrame(recordings, columns=["id", "title", "interpretation"])
```
